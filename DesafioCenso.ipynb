### Realizando o tratamento dos dados

```python
#Substituindo os Valores Incorretos Pelos maiores
adult['Tipo de Emprego'] = adult['Tipo de Emprego'].str.replace('?', 'Private', regex = True)
adult['ProfissÃ£o'] = adult['ProfissÃ£o'].str.replace('?', 'Prof-specialty', regex = True)
adult['Nacionalidade'] = adult['Nacionalidade'].str.replace('?', 'United-States', regex = True)
adult
```

```python
#Verificando se hÃ¡ dados sem preenchimento
adult.isnull().sum()
```

```python
#Selecionando apenas as colunas de interesse
adultClean = adult[['Idade','Tipo de Emprego','EducaÃ§Ã£o','ProfissÃ£o','RaÃ§a','Sexo','Horas por semana','Nacionalidade','GanhoAnual']]
adultClean
```

```python
#Ajustando os espaÃ§os entre as palavras
adultClean['Tipo de Emprego'] = adultClean['Tipo de Emprego'].apply(lambda x: x.strip())
adultClean['EducaÃ§Ã£o'] = adultClean['EducaÃ§Ã£o'].apply(lambda x: x.strip())
adultClean['ProfissÃ£o'] = adultClean['ProfissÃ£o'].apply(lambda x: x.strip())
adultClean['RaÃ§a'] = adultClean['RaÃ§a'].apply(lambda x: x.strip())
adultClean['Sexo'] = adultClean['Sexo'].apply(lambda x: x.strip())
adultClean['Nacionalidade'] = adultClean['Nacionalidade'].apply(lambda x: x.strip())
adultClean['GanhoAnual'] = adultClean['GanhoAnual'].apply(lambda x: x.strip())
adultClean
```

```python
# Verificando quantos valores possuÃ­mos por Tipo de Emprego
adultClean['Tipo de Emprego'].value_counts()
```

```python
#Agrupando os valores por categoria de EducaÃ§Ã£o
adultClean = adultClean.replace(['Preschool', '1st-4th', '5th-6th', '7th-8th' , '9th' , '10th', '11th', '12th'], 'School', regex = True) #School = PrÃ© Escola atÃ© Ensino MÃ©dio
adultClean = adultClean.replace(['HS-grad'], 'High School', regex = True)
adultClean = adultClean.replace(['Assoc-voc', 'Assoc-acdm', 'Prof-school', 'Some-college'], 'Prof. Course', regex = True)
```
