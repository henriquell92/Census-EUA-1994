### Import das bibliotecas

### Fazendo a Leitura do DataSet e inserido nome nas colunas

```python
#Lendo o data set
adult = pd.read_csv('adult.data', sep = ',')
adult
```

```python
#Nomenado as colunas
adult.columns=['Idade','Tipo de Emprego','fnlwgt','Educação','Educação-num','Estatus Civil','Profissão','Estatus de Relacionamento','Raça','Sexo','Capital Ganho','Gapital Perdido','Horas por semana','Nacionalidade','GanhoAnual']
adult
```

### Realizando o tratamento dos dados

```python
#Substituindo os Valores Incorretos Pelos maiores
adult['Tipo de Emprego'] = adult['Tipo de Emprego'].str.replace('?', 'Private', regex = True)
adult['Profissão'] = adult['Profissão'].str.replace('?', 'Prof-specialty', regex = True)
adult['Nacionalidade'] = adult['Nacionalidade'].str.replace('?', 'United-States', regex = True)
adult
```

```python
#Verificando se há dados sem preenchimento
adult.isnull().sum()
```

```python
#Selecionando apenas as colunas de interesse
adultClean = adult[['Idade','Tipo de Emprego','Educação','Profissão','Raça','Sexo','Horas por semana','Nacionalidade','GanhoAnual']]
adultClean
```

```python
#Ajustando os espaços entre as palavras
adultClean['Tipo de Emprego'] = adultClean['Tipo de Emprego'].apply(lambda x: x.strip())
adultClean['Educação'] = adultClean['Educação'].apply(lambda x: x.strip())
adultClean['Profissão'] = adultClean['Profissão'].apply(lambda x: x.strip())
adultClean['Raça'] = adultClean['Raça'].apply(lambda x: x.strip())
adultClean['Sexo'] = adultClean['Sexo'].apply(lambda x: x.strip())
adultClean['Nacionalidade'] = adultClean['Nacionalidade'].apply(lambda x: x.strip())
adultClean['GanhoAnual'] = adultClean['GanhoAnual'].apply(lambda x: x.strip())
adultClean
```

```python
# Verificando quantos valores possuímos por idade
adultClean['Idade'].value_counts()
```

```python
# Verificando quantos valores possuímos por Tipo de Emprego
adultClean['Tipo de Emprego'].value_counts()
```

```python
#Agrupando os valores por categoria de Educação
adultClean = adultClean.replace(['Preschool', '1st-4th', '5th-6th', '7th-8th' , '9th' , '10th', '11th', '12th'], 'School', regex = True) #School = Pré Escola até Ensino Médio
adultClean = adultClean.replace(['HS-grad'], 'High School', regex = True)
adultClean = adultClean.replace(['Assoc-voc', 'Assoc-acdm', 'Prof-school', 'Some-college'], 'Prof. Course', regex = True)
```

```python
# Verificando quantos valores possuímos por Educação
adultClean['Educação'].value_counts()
```

```python
# Verificando quantos valores possuímos por Profissão
adultClean['Profissão'].value_counts()
```

```python
# Verificando quantos valores possuímos por Raça
adultClean['Raça'].value_counts()
```

```python
# Verificando quantos valores possuímos por Sexo
adultClean['Sexo'].value_counts()
```

```python
# Verificando quantos valores possuímos por Horas trabalhadas na semana
adultClean['Horas por semana'].value_counts()
```

```python
# Verificando quantos valores possuímos por Nacionalidade
adultClean['Nacionalidade'].value_counts()
```

### Análise Exploratória

```python
# Relação de Homens x Mulheres que ganham  mais ou menos que 50 Mil no ano
plot = sns.countplot(adultClean['GanhoAnual'], palette='seismic', hue='Sexo', data=adultClean);
for i in plot.patches:
    plot.annotate(i.get_height(),
                 (i.get_x() + i.get_width() /2, i.get_height()),
                  ha='center', va='baseline',fontsize=12,color='black', xytext=(0,1),
                  textcoords='offset points'
                 )
```
